Setting the threshold for 7 since we dont need values above that for converion, since it is obvious that values below 7 are categorical and the rest are continous numeric values

```python
threshold = 7
```

```python
categorical_columns = unique_values[unique_values <= threshold].index.tolist()
continuous_columns = unique_values[unique_values > threshold].index.tolist()
```

#### Checking the column names that have been stored in the lists

```python
print(f'Categorical Columns : {categorical_columns}')
print()
print(f'Continuous Columns  : {continuous_columns}')
```

#### Seperating the target column 'NObeyesdad', from the categorical list, since we do not need it for conversion

```python
target_column = 'NObeyesdad'

for column in categorical_columns:
    if column == target_column:
        categorical_columns.remove(column)
        break
    
print(f'Categorical Columns : {categorical_columns}')
print()
print(f'Continuous Columns  : {continuous_columns}')
print()
print(f'Target column       : {target_column}')
```

#### Applying LabelEncoder, StandardScaler and OneHotEncoder

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder

label_encoder = LabelEncoder()
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])
    

standard_scaler = StandardScaler()
df[continuous_columns] = standard_scaler.fit_transform(df[continuous_columns])
```

#### Creating a generator and a function to train the splitted data for each classification model
##### Models used:
1. Logistic Regression
2. Decision Tree
3. Random Forest
4. Support Vector Machines
5. K-NearestNeighbor
6. Gradient Boosting
7. Multi-Layer Percepton



```python

from sklearn.metrics import classification_report, confusion_matrix


models = {
        'Logistic Regression' : LogisticRegression(max_iter = 1000),
        'Decision Tree' : DecisionTreeClassifier(),
        'Random Forest' : RandomForestClassifier(),
        'Support Vector Machines' : SVC(),
        'K-NearestNeighbor' : KNeighborsClassifier(),
        'Gradient Boosting' : GradientBoostingClassifier(),
        'Multi-Layer Percepton' : MLPClassifier(max_iter = 1000),
    }   
    

def evaluate_models_classification(X_train, X_test, y_train, y_test, models):
   
    for model_name, model in models.items():
        try:
            model.fit(X_train, y_train)
            train_score = model.score(X_train, y_train)
            test_score = model.score(X_test, y_test)
            yield model_name, train_score, test_score, None
        except Exception as e:
            yield model_name, None, None, str(e)
        print()
        
def evaluation_result(X_train, X_test, y_train, y_test, models):
    print('Evaluating the scores of each algorithm : ')
    
    best_algorithm = {
        'Name' : None,
        'Train_Score' : 0.00,
        'Test_Score' : 0.00,
        'Classification_Report' : None,
        'Confusion_Matrix' : None
    }
    
    for model_name, train_score, test_score, error_message in evaluate_models_classification(X_train, X_test, y_train, y_test, models):
        if train_score is not None: 
            print(f'{model_name} Training Score : { (train_score * 100):.2f}')
            print(f'{model_name} Test Score     : { (test_score * 100):.2f}')    
            if test_score > best_algorithm['Test_Score']:
                best_algorithm['Name'] = model_name
                best_algorithm['Train_Score'] = train_score
                best_algorithm['Test_Score'] = test_score
        else:
            print(f'Error evaluating {model_name} : {error_message}')
            
    if best_algorithm['Name']:
        print(f'Best Algorithm : {best_algorithm['Name']}')
        model = models[best_algorithm['Name']]
        model.fit(X_train, y_train)
        y_prediction = model.predict(X_test)
        best_algorithm['Classification_Report'] = classification_report(y_test, y_prediction)
        best_algorithm['Confusion_Matrix'] = confusion_matrix(y_test, y_prediction)
        print()
        print(f'Classification Report : ')
        print(best_algorithm['Classification_Report'])
        print()
        print(f'Confusion Matrix Report : ')
        print(best_algorithm['Confusion_Matrix'])
        print()
        

        
```

```python
evaluation_result(X_train, X_test, y_train, y_test, models)
```
